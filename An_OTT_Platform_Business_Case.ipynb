# Statistical summary 

The given netflix data has 8807 rows and 12 columns and has some Nan values in director,cast and country columns and data tpyes of columns are mostly obejct and one integer type of data has given.

# Missing value treatment 

By deleting or dropping

```python
df1 = df.dropna(subset=["rating"])
df1 = df1.dropna(subset=["duration"])
```

```python
df1.info()
```

By Imputation 

```python
df1["date_added"] = df1["date_added"].fillna(df1["release_year"])
df1
```

```python
df1["date_added"] = pd.to_datetime(df1["date_added"])
df1.info()

# We are changing date_added column's data type to datatime from object 
```

Filling Nan values of column cast , director and country By  "Unknown"

```python
df2 = df1.fillna("unknown")        
```

```python
df2.isna().sum() 

# Missing value and Nan value handleld 
```

## Unnesting of cast , country and listed_in columns 

```python
df_cast = df2["cast"].str.split(",").reset_index().drop( columns = ["index"])
df_cast
```

```python
df3 = df2.copy()
df3["cast"] = df_cast
df_unnest = df3.explode("cast").reset_index().drop(columns = ["index"])
df_unnest.dropna(inplace=True)

df_unnest

# unnesting of cast is completed
```

```python
df_genre = df_unnest["listed_in"].str.split(",").reset_index().drop( columns = ["index"])
df_genre
```

```python
df_unnest["listed_in"] = df_genre
df_unnest1 = df_unnest.explode("listed_in").reset_index().drop(columns = ["index"])
df_unnest1.dropna(inplace=True)

df_unnest1

# unnesting of listed_in(genre) is completed
```

```python
df_country = df_unnest1["country"].str.split(",").reset_index().drop( columns = ["index"])
df_country
```

```python
df_unnest1["country"] = df_country
df_final = df_unnest1.explode("country").reset_index().drop(columns = ["index"])
df_final.dropna(inplace=True)

df_final

#unnesting of country column is completed 
```

```python
df_final.drop_duplicates(keep="first")                # dropping all the duplicates value
```

```python
df_final.reset_index().drop(columns = ["index"])          # reseting index and dropping column index which is created by default 
```

Changing duration in to time/season and data type into Integer

```python
df_final["duration"] = df_final["duration"].str.split(" ").apply(lambda x :x[0])
df_final["duration"] = pd.to_numeric(df_final["duration"])
```

```python
df_final  

# This is our final unnested given data 
```

```python
mx = df_final["date_added"].max()
mx 
```

```python
mn = df_final["date_added"].min()
mn
```

## 2. How has the number of movies released per year changed over the last 20 years?

```python
df_last_20_year_movies = df_final.loc[(df_final["type"]=="Movie") & (df_final["release_year"] > ( df_final["release_year"].max() -20 ))  ]
df_movies_per_year = df_last_20_year_movies.groupby(["release_year"])[["show_id"]].nunique()
movies_year = df_movies_per_year.reset_index()
```

```python
plt.figure(figsize = (12,5))
sns.barplot(data = movies_year , x = "release_year"  ,  y= "show_id" )
plt.ylabel("Count")
plt.title("Movies per year")
plt.show()
```

 Here we can see that growing trend of movies per year where number of movies release per year are increasing till 2017 and after 2018 it's decreasing and there are maximum number of movies release in 2017 & 2018
